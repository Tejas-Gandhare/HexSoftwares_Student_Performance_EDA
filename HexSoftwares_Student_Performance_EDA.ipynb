```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
db=pd.read_csv("studentsPerformance.csv")
```

```python
db.head() 
# db.info()
# db.describe()
```

```python
db.shape
db.nunique
```

data cleaning

```python
db.isnull().sum()
```

```python
students = db.drop(columns=['race/ethnicity', 'parental level of education'], axis=1)
students.head()
```

relationship analysis

```python
numerical_students = students.select_dtypes(include=[np.number])
corelation = numerical_students.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.pairplot(students)
```

```python
sns.relplot(x='math score', y='reading score', hue='gender', data=students)
```

```python
sns.histplot(students['reading score'], bins=5 , kde=True)
```

```python
sns.catplot(x='math score', kind= 'box', data= students)
```
